```python
# Initial Imports


import os
import pandas as pd
from newsapi import NewsApiClient
from dotenv import load_dotenv
load_dotenv()
```

```python
api_key = os.getenv("news_api_key")
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key=api_key)
```

```python
# Fetch BTC Coin Headlines

btc_news = newsapi.get_everything(q="Bitcoin", language="en", sort_by="relevancy", from_param = "2021-12-10")
btc_news
```

```python
btc_df = pd.DataFrame.from_dict(btc_news["articles"])

btc_df
```
