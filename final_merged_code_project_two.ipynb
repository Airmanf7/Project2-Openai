```python
# Initial Imports

import flair #<------- new NLP module not used in class
from flair.data import Sentence

from flair.models import SequenceTagger
import os
import pandas as pd
from newsapi import NewsApiClient
import json
from dotenv import load_dotenv
from pathlib import Path
load_dotenv()
from path import Path
curr_date = str(pd.to_datetime('today').date()) #<------store todays date in a variable

```

```python
# Initiate a FinTech Class Sentiment Model

fintech_sentiment_model = flair.models.TextClassifier.load('en-sentiment')
```

```python
# Class Participation

chapter_1 = ""
chapter_2 = ""
chapter_3 = ""
chapter_4 = ""
chapter_5 = ""

# Combine chapters

fintech_class_movie_plot = ("One day a friendly person went for a walk on a nice day." + chapter_1 + chapter_2 + chapter_3 + chapter_4 + chapter_5)



# Run the model

results = flair.data.Sentence(fintech_class_movie_plot)
fintech_sentiment_model.predict(results)

#Define the results

sentiment = results.labels[0].value
confidence = round((results.labels[0].score) * 100, 2) #<-------Convert to percentage and round to 3 decimal places

confidence = str(confidence) #<---------- Convert to string

print(fintech_class_movie_plot)

print("The resulting sentiment is " + sentiment + " and the resulting confidence level is " + confidence + "%")
```

```python
# Initiate a Movie Sentiment Model

movie_sentiment_model = flair.models.TextClassifier.load('en-sentiment')
```

```python
# Set path to movie quotes CSV

movie_csv = Path("sentiment_moviequotes.csv")
```

```python
# Import movie quotes CSV

sentiment_moviequotes = pd.read_csv(movie_csv)
```

```python
sentiment_moviequotes
```

```python
# Initialize lists for movie quotes

movie_sentiment = []
movie_confidence = []


# Run Sentiment analysis on collected news sentences

for sentence in sentiment_moviequotes["quote"]:
    if sentence.strip() == "":
        movie_sentiment.append("")
        movie_confidence.append("")
        
    else:
        sample = flair.data.Sentence(sentence)
        movie_sentiment_model.predict(sample)
        movie_sentiment.append(sample.labels[0].value)
        movie_confidence.append(sample.labels[0].score)
```

```python
# New dataframe to collect sentiment and confidence

sentiment_moviequotes['sentiment'] = movie_sentiment
sentiment_moviequotes['confidence'] = movie_confidence
```
